```python
import numpy as np
import matplotlib.pyplot as plt

# Tentukan wilayah (domain) diagram Cartesian dan rasio lebar dan tinggi diagram
row = int(5/7*1080)
col = int(5/7*1920)

# Hitung tinggi dan lebar area yang sama
area_height = int(row / 3)
area_width = int(col)

# Tentukan persamaan matematika yang diinginkan
Rrow1 = int(0.25*row)
Rrow2 = int(0.416*row)  # Baris pertama (merah)
Rcol1 = int(0.25*col)
Rcol2 = int(0.75*col)

Wrow1 = int(0.416*row) + 1
Wrow2 = int(0.582*row)  # Baris kedua (putih)
Wcol1 = int(0.25*col)
Wcol2 = int(0.75*col)

Brow1 = int(0.582*row) + 1
Brow2 = int(0.75*row)  # Baris ketiga (biru)
Bcol1 = int(0.25*col)
Bcol2 = int(0.75*col)

Gambar = np.zeros(shape=(row, col, 3), dtype=np.uint8)  # Menggunakan np.uint8 untuk tipe data

# Mengganti warna merah pada baris pertama
for i in range(Rrow1, Rrow2+1):
    for j in range(Rcol1, Rcol2):
        Gambar[i, j, 0] = 255  # R = 255

# Mengganti warna putih pada baris kedua
for i in range(Wrow1, Wrow2+1):
    for j in range(Wcol1, Wcol2+1):
        Gambar[i, j, 0] = 255  # R = 255
        Gambar[i, j, 1] = 255  # G = 255
        Gambar[i, j, 2] = 255  # B = 255

# Mengganti warna biru pada baris ketiga
for i in range(Brow1, Brow2+1):
    for j in range(Bcol1, Bcol2+1):
        Gambar[i, j, 2] = 255  # B = 255

plt.figure()
plt.imshow(Gambar)
plt.show()
```
