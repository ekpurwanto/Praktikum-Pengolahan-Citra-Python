```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

rowmax = int(1079)
colmax = int(1919)

limit1 = int(0.2*rowmax)
limit2 = int(0.4*rowmax)
limit3 = int(0.6*rowmax)
limit4 = int(0.8*rowmax)

Gambar = np.zeros(shape=(rowmax + 1, colmax + 1 ,3), dtype=np.int16)
for i in range (0, rowmax + 1):
     for j in range(0, colmax + 1):
         if(i>=0 and i<limit1) :
             Gambar[i, j, 0] = 255
         if (i >= limit1 and i < limit2):
             Gambar[i, j, 1] = 255
         if (i >= limit2 and i < limit3):
             Gambar[i, j, 2] = 255
         if (i >= limit3 and i < limit4):
             Gambar[i, j, 0] = 255
             Gambar[i, j, 1] = 255
         if (i >= limit4 and i < rowmax):
             Gambar[i, j, 0] = 255
             Gambar[i, j, 2] = 255

#Gambar[0:limit1, 0:colmax, 0] = 255
#Gambar[limit1:limit2, 0:colmax, 1] = 255
#Gambar[limit2:limit3, 0:colmax, 2] = 255
#Gambar[limit3:limit4, 0:colmax, 0] = 255
#Gambar[limit3:limit4, 0:colmax, 1] = 255
#Gambar[limit4:rowmax, 0:colmax, 0] = 255
#Gambar[limit4:rowmax, 0:colmax, 2] = 255

plt.figure(num="")
plt.imshow(Gambar)
plt.show()
```
