Filter Rotasi Gambar 90 derajat ke kiri

```python
file_path = "/content/drive/MyDrive/slider-05.png"
pic = plt.imread(file_path)

row, col, ch = pic.shape  # Use pic instead of gambar
print(row, col, ch)

hasil_row = col
hasil_col = row
hasil_ch = ch
Gambar = np.zeros(shape=(hasil_row, hasil_col, hasil_ch), dtype=np.uint8)

# Algoritma Untuk Rotate 90 derajat
for i in range(0, hasil_row):
    for j in range(0, hasil_col):
        for k in range(0, hasil_ch):
            Gambar[i, j, k] = pic[j, i, k]

# Create a figure and display the rotated image
plt.figure(1)
plt.imshow(Gambar)
plt.title("Rotated Image")
plt.show()
```
